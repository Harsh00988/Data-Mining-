```python
#Question 4
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,auc,roc_curve
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df.diagnosis[df["diagnosis"]=="B"]=0.0
df.diagnosis[df["diagnosis"]=="M"]=1.0
df['diagnosis'] = df['diagnosis'].astype('int')
y = df['diagnosis']
df.drop("diagnosis", axis=1, inplace=True)
X = df
# error because i ran this program twice !! as we dropped the "Diagnosis"
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=13)
```

```python
Model= DecisionTreeClassifier(max_depth=5)
Model.fit(x_train, y_train)
y_pred = Model.predict(x_test)
#acc= accuracy_score(y_test,y_pred)
print("Accuracy\t:"+str(accuracy_score(y_test,y_pred)))
print("Precision\t:"+str(precision_score(y_test,y_pred)))
print("Recall\t:"+str(recall_score(y_test,y_pred)))
```

```python
prob=Model.predict_proba(x_test)
prob = prob[:,1]
fpr,tpr,_ = roc_curve(y_test, prob)
roc_auc = auc(fpr, tpr)
print("roc_auc\t:"+str(roc_auc))
```

```python
plt.figure()
lw = 2
plt.plot(
    fpr,
    tpr,
    color="darkorange",
    lw=lw,
    label="ROC curve (area = %0.2f)" % roc_auc,)
plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic example")
plt.legend(loc="lower right")
plt.show()
```

```python
Model =  RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
Model.fit(x_train, y_train)
y_pred = Model.predict(x_test)
```

```python
#acc= accuracy_score(y_test,y_pred)
print("Accuracy\t:"+str(accuracy_score(y_test,y_pred)))
print("Precision\t:"+str(precision_score(y_test,y_pred)))
print("Recall\t:"+str(recall_score(y_test,y_pred)))
prob=Model.predict_proba(x_test)
prob = prob[:,1]
fpr,tpr,_ = roc_curve(y_test, prob)
roc_auc = auc(fpr, tpr)
print("roc_auc\t:"+str(roc_auc))
```

```python
plt.figure()
lw = 2
plt.plot(
    fpr,
    tpr,
    color="darkorange",
    lw=lw,
    label="ROC curve (area = %0.2f)" % roc_auc,)
plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic example")
plt.legend(loc="lower right")
plt.show()
```
